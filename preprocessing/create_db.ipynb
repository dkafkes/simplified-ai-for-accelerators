```
#/mnt/gmps/parquet/Event0c_complete
argh = ['B|GMPSSC']
```

```
for param in argh:
  if re.search(':', param) is not None:
    param2 = param.replace(":", "")
  elif re.search('|', param) is not None:
    param2 = param.replace("|", "")
  else:
    param2 = param
  fpath = "/mnt/gmps/parquet/"+str(param)
  save_where = "/mnt/gmps/delta/just_in_case/"+str(param)
  data = spark.read.parquet(fpath)
  
  if param != 'B|GMPSSC':
    if param != 'Event0c_complete':
      data2 = data.select(col("file"), col("time"), col("value"))
      data2.write.partitionBy("file").format("delta").mode("overwrite").option("overwriteSchema", "true").save(save_where)
    else:
      data2 = data.select(col("filename").alias("file"), col("time"), col("SCmicrosec"), col("UNIXepoch"))
      data2.write.partitionBy("file").format("delta").mode("overwrite").option("overwriteSchema", "true").save(save_where)
  else:
    data2 = data.select(col("time"), col("B|GMPSSC"))
    data2.write.format("delta").mode("overwrite").option("overwriteSchema", "true").save(save_where)
  
  command = ''' CREATE TABLE {} USING DELTA LOCATION '{}';'''.format(param2, save_where)
  print(command)
  spark.sql(command)
  print("done with "+str(param))
```
